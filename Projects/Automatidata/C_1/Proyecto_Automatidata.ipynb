```python
# How many of each payment type are represented in the data?
df_automatidata['payment_type'].value_counts()
```

Código numérico que indica cómo pagó el pasajero el viaje. 

- 1= Tarjeta de crédito
- 2= Efectivo
- 3= Sin cargo

- 5= Desconocido

```python
# What is the average tip for trips paid for with credit card?
x= df_automatidata[df_automatidata['payment_type'] == 1]['tip_amount'].mean()
print(f'The average tip for trips paid for with credit card is: ${round(x,2)} dollars')

# What is the average tip for trips paid for with cash?
y= df_automatidata[df_automatidata['payment_type'] == 2]['tip_amount'].mean()
print(f'The average tip for trips paid for with cash is: ${round(y,2)} dollars')


```
